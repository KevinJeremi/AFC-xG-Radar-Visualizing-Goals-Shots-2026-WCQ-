```python
import requests
from bs4 import BeautifulSoup, Comment
import pandas as pd

url = 'https://fbref.com/en/comps/7/shooting/WCQ----AFC-M-Stats'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Cari semua komentar HTML
comments = soup.find_all(string=lambda text: isinstance(text, Comment))

# Loop semua komentar untuk cari tabel 'stats_shooting'
for comment in comments:
    if 'id="stats_shooting"' in comment:
        shooting_soup = BeautifulSoup(comment, 'html.parser')
        table = shooting_soup.find('table', {'id': 'stats_shooting'})
        if table:
            # Parsing tabel
            df = pd.read_html(str(table))[0]
            break

# Menangani kolom yang tidak diinginkan
df.columns = df.columns.droplevel()  # Hapus level yang tidak diinginkan

# Jika ada kolom 'Unnamed', ganti dengan nama yang lebih jelas
df.columns = ['Rk', 'Player', 'Pos', 'Squad', 'Age', 'Born', '90s', 'Gls', 'Sh', 'SoT', 'SoT%', 'Sh/90', 'SoT/90', 'G/Sh', 'G/SoT', 'Dist', 'PK', 'PKatt', 'Matches']


# Save the DataFrame to CSV file without sorting


print(f"Data has been saved to 'players_data.csv' with {len(df)} players")

# Display the first few rows to confirm the data
# Sort DataFrame by goals (Gls) in descending order
# Sort DataFrame by goals (Gls) in descending order
df_sorted_by_goals = df.sort_values(by='Gls', ascending=False)

# Save the sorted DataFrame to CSV
df_sorted_by_goals.to_csv('players_sorted_by_goals.csv', index=False)

print("Players sorted by most goals (all data columns preserved)")
print(f"Data has been saved to 'players_sorted_by_goals.csv'")

# Display the sorted DataFrame (showing all columns for each player)
# Displaying top 10 rows for preview
df_sorted_by_goals.head(1042)
```
